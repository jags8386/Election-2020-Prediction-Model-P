```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm



pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('../data/processed/simulation_2020.csv')
```

```python
df.head(54)
```

```python
df['other'] = 100-df.dem_perc-df.gop_perc
```

```python
def get_winner(row):
    
    bootstrap = list(np.random.choice(['D', 'R', 'O'], size = 10, p = [row['dem_perc']/100, row['gop_perc']/100,
                                                                    1-row['dem_perc']/100-row['gop_perc']/100]))
    r = bootstrap.count('R')
    d = bootstrap.count('D')
    if r > d:
        return 0
    else:
        return 1
    
    
```

```python
simulation = {}
for i in tqdm(range(100)):
    df['winner'] = df.apply(lambda x: get_winner(x), axis=1)
    df['votes'] = df['winner']*df['ev']
    simulation[i] = [df.votes.sum(), df.ev.sum() - df.votes.sum()]
    
    
    
```

```python
df_sim = pd.DataFrame.from_dict(simulation, orient='index', columns = ['D', 'R'])
```
