```python
df_2008_poll = pd.read_csv('../data/raw/all_polls_2008.csv')
```

```python
df_2008_poll['other'] = df_2008_poll['other'] + df_2008_poll['undecided']
```

```python
df_2008_poll['state'] = df_2008_poll['state'].str.replace('--', 'National')
```

```python
df_2008_poll.columns
```

```python
df_2008_poll['number.of.observations'] = df_2008_poll['number.of.observations'].fillna(
    df_2008_poll.groupby('state')['number.of.observations'].transform('mean'))
```

```python
df_2008_poll.head()
```

```python
df_2008_poll['total_sample'] = df_2008_poll.groupby(['state'])['number.of.observations'].transform('sum')
```

```python
df_2008_poll['weight'] = df_2008_poll['number.of.observations'] / df_2008_poll['total_sample']
```

```python
df_2008_poll['perc_dem'] = df_2008_poll['obama'] * df_2008_poll['weight']
df_2008_poll['perc_gop'] = df_2008_poll['mccain'] * df_2008_poll['weight']
```

```python
df_2008_poll_sum = df_2008_poll.groupby('state')[['number.of.observations',
                                                 'perc_dem', 'perc_gop']].sum().reset_index()
df_2008_poll_sum['year'] = 2008
```

```python
df_2008_poll_sum.head()
```

```python
df_2012_poll.isna().sum()
```

```python
df_2012_poll['state'] = df_2012_poll['state'].str.replace('--', 'National')
df_2012_poll['number.of.observations'] = df_2012_poll['number.of.observations'].fillna(
    df_2012_poll.groupby('state')['number.of.observations'].transform('mean'))

df_2012_poll['total_sample'] = df_2012_poll.groupby(['state'])['number.of.observations'].transform('sum')
df_2012_poll['weight'] = df_2012_poll['number.of.observations'] / df_2012_poll['total_sample']

df_2012_poll['perc_dem'] = df_2012_poll['obama'] * df_2012_poll['weight']
df_2012_poll['perc_gop'] = df_2012_poll['romney'] * df_2012_poll['weight']

df_2012_poll_sum = df_2012_poll.groupby('state')[['number.of.observations',
                                                 'perc_dem', 'perc_gop']].sum().reset_index()


df_2012_poll_sum['year'] = 2012
```
